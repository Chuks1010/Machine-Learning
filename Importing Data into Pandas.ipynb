## Import Data into Pandas with Python

One of the first operations in data analysis is importing data into pandas dataframe. There are a number of format that can be imported into pandas dataframe. Let us consider flat file format. I will demostrate using csv, txt, excel and json 

CSV Files

CSV is one of the most commonly used data format. It stands for comma seperated values and as the name implies such file uses comma to seperate values. Each line of file is a data record ie a record of an instance. To import the csv file into a pandas dataframe we use the pd.read_csv() function passing the file name as parameter.

```python
import pandas as pd
loans = pd.read_csv('loans_2007.csv')
print(loans.shape)
loans.head()
```

MicroSoft Excel is a spreadsheet developed by MicroSoft and popularly used in the business world. Its a proprietary file format with extension xlxs.It is an excelence tool for data analysts. The pd.ExcelFile() imports ad store the Excel file into a pandas dataframe. it is suitable if the excel file is made up of one of more than one sheet. You can also use pd.read_excel() function if the file has only one sheet.

```python
raw = pd.ExcelFile('raw_data.xlsx')
print(raw.sheet_names)
```

```python
new_cust = pd.read_excel('new_cust1.xlsx')
print(new_cust.shape)
new_cust.head()
```

Text File

Text files have file extension of txt. They are useful in storing information in plain text which maybe in tabular or non-tabular format. They can be opened with pd.read_csv() or pd.read_table().

```python
pd.read_table('rfm_xmas19.txt')
```

```python
data = pd.read_table('testme.txt')
print(data)
```

Database

Organizational data are usually stored in relational or non-relational databases for structured data or unstructured data respectively. These data can be imported from a database into a pandas dataframe. To do this the database has to be import first and connected established in order to pull the data from the database into pandas  dataframe. The server, user and database information are required to establish this connection. Once the connection is established, you can run the SQL query to extract the relevant information and store it in the dataframe for further analysis 

Universal Resource Locator (URL)

Data can be available on a website.To access and downlaod it we need to import urllib package, specify the url and download it and display it in andas dataframe

```python
import urllib
from urllib.request import urlretrieve
url = 'http://winterolympicsmedals.com/medals.csv'
urlretrieve(url, 'medals.csv')

web_data = pd.read_csv('medals.csv')
print(web_data.shape)
web_data.head() 
```
